读取excle

参考：https://zhuanlan.zhihu.com/p/142972462

```python
# sheet_name 读取那个sheet
    # 默认 sheet_name=0 读取第一个sheet
    # sheet_name=[0,'sheet_name'] 读取第一个 sheet 和指定 sheet_name，会读取到一个 orderedDict 中
    # sheet_name=None 读取所有sheet
# header 表头，默认0，表示第一行
    # header = 1 第二行是表头，第一行就不要了
    # header = [1,2,3] 第234行是表头
    # header =None 没有表头
# index_col 默认 None，表示使用第几列为索引
    #index_col=None 不适用数据列作为索引，自增0开始
    #index_col=0 使用数据第一列作为索引
    #index_col=[0,2] 使用指定数据列的 整数列为索引列
pd.read_excel(
    './data/team.xlsx'
    ,sheet_name=0
    ,header=0
    ,index_col=0
    )
```

# 读取多个或所有sheet

sheet_name 为 list

```python
# sheet_name 传入 list 为 读取多个sheet
pd.read_excel(
    './data/team.xlsx'
    ,sheet_name=[0,'info']
    )
```

sheet_name=None

```python
multi_df=pd.read_excel(
    './data/team.xlsx'
    ,sheet_name=None
    )
multi_df
```

循环多个 sheet

```python
for pre in multi_df:
    multi_df[pre]
```

# 保存excle

```python
# to_excel excel不存在创建，存在覆盖
# index=False 不写入索引
df=pd.read_excel('./data/team.xlsx')
df.to_excel(
    'test.xlsx'
    ,sheet_name='hello'
    ,index=False
)
```
