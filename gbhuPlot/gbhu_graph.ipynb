# networkx

参考：官网https://www.osgeo.cn/networkx/tutorial.html

```python
import networkx as nx
print(nx.__version__)
```

```python
# 创建一个简单的图
G = nx.Graph()
# 添加节点
G.add_nodes_from([1, 2, 3, 4])
# 添加边
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])
```

```python
# 创建一个简单的图
G = nx.Graph()
# 添加节点
G.add_nodes_from(['1', '2', '3', '4'])
# 添加边
G.add_edges_from([('1', '2'), ('2', '3'), ('3', '4'), ('4', '1')])
```

## networkx+matplotlib 绘图

效果不咋地

```python
import matplotlib.pyplot as plt
nx.draw(G, with_labels=True, font_weight='bold')
plt.show()
```

## networkx+pyvis 绘图

参考：https://pyvis.readthedocs.io/en/latest/tutorial.html?highlight=networkx#networkx-integration

```python
from pyvis.network import Network
import pyvis
print(pyvis.__version__)
```

```python
nt = Network('500px', '500px')
nt.from_nx(G)
# notebook=False 必须，不然会报错，三方包的问题
nt.show("g.html", notebook=False)
# 保存到本地
nt.save_graph('test.html')
```
