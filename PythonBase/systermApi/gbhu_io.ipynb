循环读取每一行

```python
os.path.join(os.getcwd(),'gbhu_os.ipynb')
with open(
    os.path.join(os.getcwd(),'gbhu_os.ipynb'),
    'r',
    encoding='utf-8'
)as fp:
    for line in fp:
        print(line)
```

循环读取每一行，带行号

```python
os.path.join(os.getcwd(),'gbhu_os.ipynb')
with open(
    os.path.join(os.getcwd(),'gbhu_os.ipynb'),
    'r',
    encoding='utf-8'
)as fp:
    #start=i 行号从几开始
    for line_nunmber,line in enumerate(fp, start=1):
        print(line_nunmber,line)
```
